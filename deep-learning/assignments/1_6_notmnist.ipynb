Train a simple model on this data using 50, 100, 1000 and 5000 training samples. Hint: you can use the LogisticRegression model from sklearn.linear_model.

```python
import pickle
from sklearn import cross_validation
from sklearn import linear_model

pickle_file = 'notMNIST.pickle'
f = open(pickle_file, 'rb')
pdata = pickle.load(f)

train_dataset = pdata['train_dataset']
train_labels= pdata['train_labels']
valid_dataset = pdata['valid_dataset']
valid_labels=pdata['valid_labels']
test_dataset=pdata['test_dataset']
test_labels=pdata['test_labels']

train_dataset_lr = train_dataset.reshape( train_dataset.shape[0], train_dataset.shape[1]*train_dataset.shape[2])
test_dataset_lr = test_dataset.reshape( test_dataset.shape[0], test_dataset.shape[1]*test_dataset.shape[2])
print(train_dataset_lr.shape)

train_set_size = 1000

train_dataset_lr = train_dataset_lr[0:train_set_size,:]
print(train_dataset_lr.shape)
train_labels_lr = train_labels[0:train_set_size]

# Initialize the algorithm class
alg = linear_model.LogisticRegression(random_state=1)

# Train the algorithm using all the training data
alg.fit(train_dataset_lr,train_labels_lr)

score = alg.score(test_dataset_lr,test_labels)

print(score)
# Make predictions using the test set.
#predictions = alg.predict(test_dataset)
```
